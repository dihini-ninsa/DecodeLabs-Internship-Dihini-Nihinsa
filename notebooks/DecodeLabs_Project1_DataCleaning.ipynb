***Phase 1: Strategic Imputation (Handle Missing Values)***

```python
from google.colab import drive
drive.mount('/content/drive')

import pandas as pd

df = pd.read_excel('/content/drive/MyDrive/Datasets/Decodelab Internship/Dataset for Data Analytics.xlsx')
df['CouponCode'] = df['CouponCode'].fillna('None')
```

***Phase 2: The Integrity Audit (Duplicates)***

```python
# Check
print("Duplicate OrderIDs:", df['OrderID'].duplicated().sum())  # Should be 0

# Remove (if needed)
df = df.drop_duplicates(subset='OrderID', keep='first')
```

***Phase 3: Correct Data Formats***

```python
# Ensure date column is datetime
df['Date'] = pd.to_datetime(df['Date'])

# When saving, the format is preserved
```

***Numbers — 2 decimal places***

```python
text_cols = ['Product', 'PaymentMethod', 'OrderStatus', 'ReferralSource', 'ShippingAddress']
for col in text_cols:
    df[col] = df[col].str.strip().str.title()
```

***Save the Clean File***

```python
df.to_excel('/content/drive/MyDrive/Datasets/Decodelab Internship/Cleaned_Dataset.xlsx', index=False)
print("Done! Rows:", len(df))
```

***Verification Gate (to pass to Project 2)***

```python
# Proof 1: Zero duplicate OrderIDs
assert df['OrderID'].duplicated().sum() == 0, "Duplicates found!"

# Proof 2: Zero missing values (after cleaning)
assert df.isnull().sum().sum() == 0, "Nulls found!"

print("Verification passed. Zero duplicate IDs. Zero nulls.")
```
